```python
import pandas as pd
import plotly.express as px
```

```python
#Leyendo archivo
data = pd.read_csv('bikes.csv', sep = ";", encoding='utf8')
data
```

```python
#Graficando cantidad de bicicletas retiradas por dias para Berr 1
fig = px.line(data, x = "Date", y = "Berri 1")
fig.show()
```

```python
#Convirtiendo las fechas(str) a tipo fecha
data['Date'] = pd.to_datetime(data['Date'], dayfirst = True)
data
```

```python
#Graficando cantidad de bicicletas retiradas por dias para Berr 1 con el arreglo de fechas
fig = px.line(data, x = "Date", y = "Berri 1", title = "Cantidad de bicicletas retiradas")
fig.show()
```

```python
#Agrupando por dia
data_by_date = data.groupby(['Date']).sum()
data_by_date
```

```python
#Haciendo un resample para agrupar por mes
data_by_month = data_by_date.resample("M").sum()

#Cambiando formato de fecha para q solo aparezca el mes
data_by_month.index = data_by_month.index.strftime('%B')
data_by_month
```

```python
#Graficando cantidad de bicicletas retiradas por mes
fig = px.line(data_by_month, data_by_month.index, 
              y = ['Berri 1', 'Côte-Sainte-Catherine','Maisonneuve 1', 
                   'Maisonneuve 2', 'du Parc', 'Pierre-Dupuy', 'Rachel1'], 
              title = "Cantidad de trayectos por mes")
fig.show()
```

```python
#Graficando cantidad de bicicletas retiradas por mes
fig = px.bar(data_by_month, data_by_month.index, 
              y = ['Berri 1', 'Côte-Sainte-Catherine','Maisonneuve 1', 
                   'Maisonneuve 2', 'du Parc', 'Pierre-Dupuy', 'Rachel1'], 
              title = "Cantidad de trayectos por mes")
fig.show()
```

```python
#Convirtiendo los dias a semanas
data_weekday =  data_by_date
data_weekday['Weekday'] = data_weekday.index.strftime('%A')

#Agregando promedio y desviacion estandar
data_info = data_by_date.groupby(['Weekday']).agg(['mean', 'std'])
data_info
```

```python
#Guardando archivos
#JSON
data_by_month.to_json('bikesByMonth.json', compression = 'infer', index = 'true')
data_info.to_json('bikesByWeekday.json',compression = 'infer', index = 'true')

#CSV
data_by_date.to_csv("data_update.csv")
```

```python
#Probando boxplot

#Agregando columnas de suma total y dias de la semana para cada dia
data_week_total = data_by_date.groupby(['Date']).sum()
data_week_total['Total'] = data_week_total.agg("sum", axis="columns")
data_week_total['Weekday'] = data_week_total.index.strftime('%A')

#ploteando
fig = px.box(data_week_total, x="Weekday", y="Total")
fig.show()
```
